````python
# Q6-2 Grading Tag:


def score_word(word: str) -> int:
    '''Scores a Scrabble word.  Argument is a string 2-7 characters including alphabet and * for blank'''


    score = 0


    # Score dictionary copied below, "*" added to end of dictionary with value of 0
    scores = { "a": 1, "c": 3, "b": 3, "e": 1,  "d": 2, "g": 2,
            "f": 4, "i": 1, "h": 4, "k": 5,  "j": 8, "m": 3,
            "l": 1, "o": 1, "n": 1, "q": 10, "p": 3, "s": 1,
            "r": 1, "u": 1, "t": 1, "w": 4,  "v": 4, "y": 4,
            "x": 8, "z": 10, "*": 0 }
    
    for letter in word.lower():
        score += scores[letter]

    return score
    


def validate_rack(rack: str) -> bool:
    '''Returns True if rack is Valid and False if rack is invalid.  Validty is between 2-7 characters, alphabet characters only, and wild cards * '''
    valid_chars = "abcdefghijklmnopqrstuvwxyz*"

    if 1 < len(rack) < 8:
        for letter in rack:
            if letter.lower() in valid_chars:
                continue
            else:
                return False
    else:
        return False
    
    return True


def search_range(letters: str) -> tuple:                # I wrote this function thinking I would narrow down the search range and speed things up
                                                        # This ended up becoming a permutations approach, though, so I didn't use it in the final code.
    '''returns a search index range as a tuple
        (start, stop) for the provided string'''

    start = 0
    stop = 0

    for i,word in enumerate(WORDS):                     # This will provide the starting index for the search range
        if word[:len(letters)] == letters.upper():
            start = i
            break
        
    
    for i,word in enumerate(reversed(WORDS)):           # This will provide the stopping index for the search range
        if word[:len(letters)] == letters.upper():
            stop = len(WORDS)-i-1
            break
        
    return (start,stop)     # For now, if word is not found (0,0) is returned


def run_scrabble(rack: str, words: list[str]) -> dict[str, int]:
    '''Returns a dictionary of possible scrabble words and their associated score'''
    valid_rack = validate_rack(rack)                    # Validate the rack before doing any calculations

    result_dict = {}                                    # Create empty dictioanry for storing results

    if valid_rack:

        for word in words:
            
            rack1 = rack.upper()                            # Temporary rack used to try out each word
            word_score = ""                                 # Empty string used to store word for scoring purposes

            for i,letter in enumerate(word):                # enumeration needed for replacing letters with *'s
                if letter in rack1:
                    rack1 = rack1.replace(letter, "", 1)    # Removed used letter
                    word_score += letter                    # Add Letter to scoring word
                elif "*" in rack1:
                    rack1 = rack1.replace("*", "", 1)       # remove wild card from rack
                    word_score += "*"                       # Add * to scoring word
                else:
                    break

            else:
                result_dict[word] = score_word(word_score)  # Add word and score to results, if loop completes

    else:

        return print("You did not enter a valid rack, please enter a rack betwen 2-7 characters with alphabetical characters and wildcards '*' only")


    return dict(sorted(result_dict.items(), key = lambda item: item[1], reverse = True))    #Returns results sorted by largest score




if __name__ == "__main__":
    # The condition above checks if this script is being run directly or being imported. More info:
    # https://docs.python.org/3/library/__main__.html
    # https://realpython.com/if-name-main-python/#in-short-it-allows-you-to-execute-code-when-the-file-runs-as-a-script-but-not-when-its-imported-as-a-module

    rack = "EXAMPLE"  # feel free to use input() to get the rack from the user.
    # TODO: Call the functions above to complete the scrabble solver
````
